# Web Scraping

Este notebook foi desenvolvido com objetivo de estudar Web Scraping com Python, baseado na formação "Python para Data Science" da Alura, obtendo dados de https://alura-site-scraping.herokuapp.com/index.php

## Obtendo o html

```python
from urllib.request import Request, urlopen

url = 'https://alura-site-scraping.herokuapp.com/index.php'  # Url que será acessado

response = urlopen(url)  # Guarda a resposta ao request de acesso da página

html = response.read()  # Abre o html recebido

html
```

```python
type(html) #tipo de dado do html
```

### Tratamento do html

```python
#Convertendo para string
html = html.decode('utf-8')

type(html)
```

```python
html #caracteres que não estavam codificados agora aparecem
```

```python
# Tratando as strings

html = html.split()  # remove as marcações html como \n

html = ' '.join(html).replace('> <', '><')  # remove espaços entre tags

html
```

## Interpretando, buscando e extraindo informação

```python
from bs4 import BeautifulSoup
```

```python
#Objeto BeautifulSoup
soup = BeautifulSoup(html, 'html.parser')

print(soup.prettify())  # Exibe o html "bonitinho"
```

### Acesso a tags
O BeautifulSoup lê o html de cima para baixo, o acesso as tags é feito como que se fossem atributos do objeto soup:
    
    soup.html.head.title
    
Acessando sempre a primeira tag encontrada

### Acessando conteúdo das tags
O método .get_text() traz o conteúdo da tag:
    
    soup.title.get_text()

### Acessando atributos das tags
Também é possível acessar atributos da tag:
    
   soup.img.attrs; soup.img.attrs.keys(); soup.img.attrs.values() [o .attrs retorna um dicionário do tipo {'Atributo': 'Valor'}]
    
   soup.img['src']; soup.get('src') [acessa diretamente o atributo]

### Pesquisa com BeautifulSoup

Com .find('tag') e .findAll('tag') [ou soup('tag')] encontramos a primeira, ou todas, respectivamente, as tags passadas como parâmetro

## Capturando dados dos Anuncios

Observando a estrutura da página, as informações dos anuncios estão dentro de cards, em divs da classes 'well card'

### Para a primeira página

```python
import pandas as pd
from urllib.request import urlretrieve
```

```python
cards = []  # Lista que armazenará todos os dicionários 'card'

containers = soup.find('div', {'id': 'container-cards'}).findAll('div', {'class': 'well card'})

counter = 0

for card in containers:
    anuncio = {}  # Dicionário para cada um dos anuncios
    
    infos = card.findAll('p')  # Acessa as tags <p> onde estão as informações dos cards
        
    for info in infos:  # Varre cada uma das <p> criando pares key:value e adicionando ao dicionario card
        anuncio[info.get('class')[0].split('-')[-1]] = info.get_text()
    
    acessorios = []
    itens = card.find('div', {'class': 'body-card'}).ul.findAll('li')  # Acessa as tags <li> do card onde estão os acessórios do veículo
    itens.pop()
    for item in itens: # Varre cada uma das <li> criando uma lista com os acessórios descritos no anúncio
        acessorios.append(item.get_text().replace('► ', ''))    
    anuncio['acessorios'] = acessorios  #  Lista dos acessórios
    
    cards.append(anuncio) # Salva anuncio capturado na lista
    
    
    #Salva imagem
    image = card.img.get('src')
    urlretrieve(image, 'images/img-'+str(counter)+'.png')
    
    counter += 1

cars_dataset = pd.DataFrame(cards)#, orient = 'index').T  #  Dataset com as infos extraidas  
```

### Para todo o site

```python
n_paginas = int(soup.find('span', class_='info-pages').get_text().split()[-1])
```

```python
n_paginas
```

```python
cards = []  # Lista que armazenará todos os dicionários 'card'

counter = 0

for pagina in range(n_paginas):
    
    url = 'https://alura-site-scraping.herokuapp.com/index.php?page='+str(pagina+1)  # Url que será acessado

    response = urlopen(url)  # Guarda a resposta ao request de acesso da página

    html = response.read()  # Abre o html recebido
    html = html.decode('utf-8')

    soup = BeautifulSoup(html, 'html.parser')

    containers = soup.find('div', {'id': 'container-cards'}).findAll('div', {'class': 'well card'})

    for card in containers:
        anuncio = {}  # Dicionário para cada um dos anuncios

        infos = card.findAll('p')  # Acessa as tags <p> onde estão as informações dos cards

        for info in infos:  # Varre cada uma das <p> criando pares key:value e adicionando ao dicionario card
            anuncio[info.get('class')[0].split('-')[-1]] = info.get_text()

        acessorios = []
        itens = card.find('div', {'class': 'body-card'}).ul.findAll('li')  # Acessa as tags <li> do card onde estão os acessórios do veículo
        itens.pop()
        for item in itens: # Varre cada uma das <li> criando uma lista com os acessórios descritos no anúncio
            acessorios.append(item.get_text().replace('► ', ''))    
        anuncio['acessorios'] = acessorios  #  Lista dos acessórios

        cards.append(anuncio) # Salva anuncio capturado na lista


        #Salva imagem
        image = card.img.get('src')
        urlretrieve(image, 'images/img-'+str(counter)+'.png')

        counter += 1

cars_dataset = pd.DataFrame(cards)#, orient = 'index').T  #  Dataset com as infos extraidas
```
