O relatório de análise a seguir foi produzida com objetivos de estudar Análise de Dados com Python e Pandas, a base de dados "aluguel.csv" foi obtida a partir do curso "Python para Data Science" da Alura.

## Importação de dados

```python
import pandas as pd
```

```python
aluguel_dataset = pd.read_csv('data/aluguel.csv', sep = ';')

aluguel_dataset
```

## Informações sobre a base de dados

```python
aluguel_dataset.info()
```

```python
n_linhas = aluguel_dataset.shape[0]
n_colunas = aluguel_dataset.shape[1]

print(f'A base de dados possui { n_linhas} registros de imóveis, com {n_colunas} variáveis caracterizando-os')
```

```python
tipagem_dos_dados = pd.DataFrame(aluguel_dataset.dtypes, columns = ['Tipos de Dados'])
tipagem_dos_dados.columns.name = 'Variáveis'

tipagem_dos_dados
```

### Tipos de imóveis na base

```python
imoveis = aluguel_dataset[['Tipo']] #salva apenas a coluna 'Tipo' como um dataframe do pandas

imoveis.drop_duplicates(inplace = True) #remove as duplicatas dos imoveis

imoveis.index = range(imoveis.shape[0]) #altera o indice do dataframe imoveis

imoveis
```

## Imóveis Residenciais

Os imóveis considerados residenciais são: Quitinete, Casa, Apartamento, Casa de Condomínio e Casa de Vila

```python
imoveis_residenciais = ['Quitinete', 'Casa', 'Apartamento', 'Casa de Condomínio', 'Casa de Vila']
```

```python
select_residenciais = aluguel_dataset['Tipo'].isin(imoveis_residenciais) #cria uma Series booleana de comparação entre a coluna tipo e os valores na lista passada

select_residenciais
```

```python
aluguel_residencial = aluguel_dataset[select_residenciais] #cria um novo dataframe a partir da Series booleana

aluguel_residencial.index = range(aluguel_residencial.shape[0]) #refaz a indexação

aluguel_residencial
```

### Exportando dados

```python
aluguel_residencial.to_csv('data/outputs/alugues_residenciais.csv', sep = ';', index = False) #Salva o dataframe como um arquivo csv, sem o index 
```

## Extraindo informações de "alugueis_residenciais.csv"

### Apartamentos

```python
select_tipo_apartamento = aluguel_residencial['Tipo'] == 'Apartamento'
n_apartamentos = aluguel_residencial[select_tipo_apartamento].shape[0] # "Contagem de apartamentos"
media_valor_apartamentos = aluguel_residencial.Valor[select_tipo_apartamento].mean().round(2)

print(f'Nossa base possui {n_apartamentos}, com alugueis de R${media_valor_apartamentos} em média')
```

### Casas, Casas de Condominio, Casas de Vila

```python
select_tipo_casas = aluguel_residencial['Tipo'].isin(['Casa', 'Casa de Condomínio', 'Casa de Vila'])
n_casas = aluguel_residencial[select_tipo_casas].shape[0]
media_valor_casas = aluguel_residencial.Valor[select_tipo_casas].mean().round(2)

print(f'Temos {n_casas} em nossa base, com um valor de aluguel de R${media_valor_casas} em média')
```

### Imóveis com área entre 60 e 100 metros quadrados

```python
select_area_60_100 = (aluguel_residencial['Area'] >= 60) & (aluguel_residencial['Area'] <= 100)
n_imoveis_area_60_100 = aluguel_residencial[select_area_60_100].shape[0]

print(f'{n_imoveis_area_60_100} possuem área entre 60 e 100 m²')
```

### Imóveis com ao menos 4 quartos e possuem aluguel menor que R$2.000,00

```python
select_4quartosMais_menos2K = (aluguel_residencial['Quartos'] >= 4) & (aluguel_residencial['Valor'] < 2000)
n_imoveis_4quartosMais_menos2K = aluguel_residencial[select_4quartosMais_menos2K].shape[0]

print(f'Dos imóveis, {n_imoveis_4quartosMais_menos2K} possuem ao menos 4 quartos e têm aluguel menor que R$2000')
```

## Tratando dados

### Dados faltantes

Pelas informações de "aluguel_residencial"

```python
aluguel_residencial.info()
```

As colunas 'Valor', 'Condominio' e 'IPTU' apresentam valores nulos

#### Tratando nulos da coluna 'Valor'

Neste caso, não nos interessam registros cujo valor do aluguel seja nulo.

```python
n_valor_nulos = aluguel_residencial[aluguel_residencial['Valor'].isnull()].shape[0]

print(f'Temos {n_valor_nulos} linhas com aluguel nulo na coluna "Valor"')
```

```python
#removendo valores nulos de 'Valor'
aluguel_residencial.dropna(subset = ['Valor'], inplace = True)

aluguel_residencial.info()
```

#### Tratando nulos de Condominio e IPTU

No caso do condominio, nem todo imóvel tem condomínio. Então, não seria prudente remover todos os registros 
em que o 'Condominio' == NaN. Aqui, assume-se que se o imóvel é um apartamento, ele deveria ter um valor de condominio

```python
#identificando apartamentos com 'Condominio' == NaN

select_apartamento_sCondominio = (aluguel_residencial['Tipo']=='Apartamento') & (aluguel_residencial['Condominio'].isnull())

print(f'Temos {aluguel_residencial[select_apartamento_sCondominio].shape[0]} apartamentos, sem valor de condominio')
```

```python
#removendo a selecao identificada
aluguel_residencial2 = aluguel_residencial[~select_apartamento_sCondominio] #criando um novo dataframe, com o inverso da selecão
```

```python
aluguel_residencial2.info()
```

Agora, todos os valores nulos da base podem ser substituidos por 0

```python
#substituindo NaN em 'Condomínio' e 'IPTU' por 0

aluguel_residencial2 = aluguel_residencial2.fillna({'Condominio': 0, 'IPTU': 0})

aluguel_residencial2.info()
```

```python
#salvando csv com dataframe tratado
aluguel_residencial2.to_csv('data/outputs/aluguel_residencial_V2.csv', sep = ';', index = False)
```

## Novas Variaveis

### Valor bruto

Valor total pago mensalmente ao alugar o imóvel: 'Valor' + 'Condominio' + 'IPTU'

```python
aluguel_residencial2['Valor Bruto'] = aluguel_residencial2['Valor'] + aluguel_residencial2['Condominio'] + aluguel_residencial2['IPTU']

aluguel_residencial2
```

### Valor por metro quadrado

O valor de aluguel por metro quadrado

```python
aluguel_residencial2['Valor por m2'] = aluguel_residencial2['Valor']/aluguel_residencial['Area']

aluguel_residencial2['Valor por m2'] = aluguel_residencial2['Valor por m2'].round(2)

aluguel_residencial2
```

### Valor Bruto por metro quadrado

```python
aluguel_residencial2['Valor Bruto por m2'] = (aluguel_residencial2['Valor Bruto']/aluguel_residencial2['Area']).round(2)

aluguel_residencial2
```

### Grupo

Agrupar imóveis, por tipo, em 'Casa' ou 'Apartamento'

```python
aluguel_residencial2['Grupo'] = aluguel_residencial2['Tipo'].apply(
    lambda x: 'Casa' if x in ['Casa', 'Casa de Condomínio', 'Casa de Vila'] else 'Apartamento')

aluguel_residencial2
```

No entanto, 'Valor Bruto' e 'Valor bruto por m2' podem não refletir muito bem a realidade, por conta dos valores nulos que substituimos

```python
aluguel_residencial3 = aluguel_residencial2.drop(['Valor Bruto', 'Valor Bruto por m2'], axis = 1) #removendo colunas

aluguel_residencial3
```

```python
aluguel_residencial3.to_csv('data/outputs/aluguel_residencial_V3.csv', sep = ';', index = False)
```

## Agrupamento de dados

```python
select_bairros = aluguel_residencial3['Bairro'].isin(['Barra da Tijuca', 'Flamengo', 'Botafogo', 'Copcabana', 'Ipanema', 'Tijuca'])

aluguel_bairros1 = aluguel_residencial3[select_bairros]

aluguel_bairros1
```

```python
aluguel_bairros_grupo1 = aluguel_bairros1.groupby('Bairro') #agurpa pelos bairros selecionados

aluguel_bairros_grupo1 = aluguel_bairros_grupo1[['Valor', 'Condominio']].mean().round(2) #adiciona a média de valor e de condominio como variaveis

aluguel_bairros_grupo1
```

## Plotando informações

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.rc('figure', figsize = (20, 10))
```

```python

x = aluguel_bairros_grupo1.index
y = aluguel_bairros_grupo1['Valor']

plt.bar(x, y, color = 'blue')

plt.xlabel('Bairro')
plt.ylabel('Valor Médio')
plt.title('Aluguel médio por Bairro')
```

## Outliers

```python
#Qartis do Valor de aluguel
valor_aluguel = aluguel_residencial.Valor

Q1 = valor_aluguel.quantile(.25) #1º quartil
Q3 = valor_aluguel.quantile(.75) #3º quartil
IIQ = Q3 - Q1 #intervalo interquartis

limite_inferior = Q1 - 1.5*IIQ
limite_superior = Q3 + 1.5*IIQ
```

```python
#seleção entre os limites

select_valor_entre_limites = (valor_aluguel >= limite_inferior) & (valor_aluguel <= limite_superior)
```

```python
dataset_sOutliers = aluguel_residencial[select_valor_entre_limites]

dataset_sOutliers
```
