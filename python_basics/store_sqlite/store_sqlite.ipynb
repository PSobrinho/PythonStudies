```python
import sqlite3 as sql

conn = sql.connect("store.db")
cur = conn.cursor()
    
```

```python
def create_table(): #criação de tabela como função
    cur.execute("CREATE TABLE IF NOT EXISTS produtos(id INTEGER PRIMARY KEY NOT NULL, prod_name TEXT, price REAL)")

def insert_data(recset): #inserção de dados como função
    sql_insert = "INSERT INTO produtos VALUES (?, ?, ?)"
    
    for rec in recset:
        cur.execute(sql_insert, rec)
    
    conn.commit()
        
def consult_all_data(): #consulta via função
    cur.execute("SELECT * FROM produtos")
    data = cur.fetchall()
    
    for row in data:
        print("Produto Id: %d, Produto: %s, Preço: %f" %row)
```
