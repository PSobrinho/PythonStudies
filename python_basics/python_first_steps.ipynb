A função a seguir recebe uma entrada de dado do usuário, e em seguida retorna uma saudação utilizando este nome

```python
def boasVindas():
  nome = input('Qual o seu nome? ')
  print(f'Seja bem-vinde {nome}')
```

```python
boasVindas()
```

Esta função verifica se o usuário está apto a comprar bebidas alcoólicas

```python
def comprarBebida():
  idade = input('Qual a sua idade? ')
  idade = int(idade)
  if idade >= 18:
    print('Maior idade confirmada. Realizando pedido...')
  else:
    print('Não é permitida venda de bebidas alcoólicas para menores. Pedido cancelado.')
```

```python
comprarBebida()
```

```python
comprarBebida()
```

Calculo de velocidade média, em metros por segundo [m/s]

```python
def velocidadeMedia():
  metros = input('Entre com a distância em metros [m]: ')
  segundos = input('Entre com o tempo em segundos [s]')
  velocidade = float(metros)/float(segundos)
  print(f'A velocidade média é de {velocidade} m/s')
```

```python
velocidadeMedia()
```

Verificar cada a aprovação de cada aluno, a partir de uma lista de médias (Operadores de comparação: ==, !=, <=,>=, <, >)


```python
from random import randrange

notas = [] #lista de notas

for nota in range(8):
  nota = randrange(0,11)
  notas.append(nota)

notas
#Com list comprehensions, a lista poderia ser criada como: notas = [randrange(0, 11) for nota in range(8)]
```

```python
aprovacoes = [] #lista para armazenar situações de cada nota
```

```python
def isAprovado(notas, aprovacoes):
  for nota in notas:
    if nota >= 6:
      aprovacoes.append(True)
    else:
      aprovacoes.append(False)

  for aprovacao in aprovacoes:
    if aprovacao:
      print('Aluno aprovado')
    else:
      print('Aluno reprovado')
```

```python
isAprovado(notas, aprovacoes)
```

Gráfico de notas

```python
import matplotlib.pyplot as plt

x = list(range(1, 9))
y = notas

plt.plot(x, y, marker = 'o')
plt.title('Notas')
plt.xlabel('Provas')
plt.ylabel('Notas')
plt.show()
```
