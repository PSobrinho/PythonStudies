#Queries

```python
select = cars_dataset2.Ano >= 2000

cars_dataset2[select] #seleciona apenos as linhas cujo ano é >=2000
```

```python
cars_dataset2[(cars_dataset2.Ano >= 2000) & (cars_dataset2.Zero_km == True)] #seleciona todos as linhas cujo ano é >=2000 e é um carro zero
```

```python
cars_dataset2.query('Ano >= 2000 and Zero_km == True and Valor < 100000') #querie construida com o método .query()
```

```python
list(cars_dataset2.iterrows()) #exemploficando o resultado do iterrows
```

```python
for index, row in cars_dataset2.iterrows(): #itera cada uma das linhas calculando uma nova coluna
  if(2021 - row['Ano'] != 0):
    cars_dataset2.loc[index, 'km media'] = row['Quilometragem']/(2021 - row['Ano'])
  else:
    cars_dataset2.loc[index, 'km media'] = 0

cars_dataset2
```
