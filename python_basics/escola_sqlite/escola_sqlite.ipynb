```python
import os #verifica se existe já existe um arquivo "escola.db", e o exclui caso sim.
os.remove("escola.db") if os.path.exists("escola.db") else None
```

```python
import sqlite3 #importa o SQLite
```

```python
con = sqlite3.connect("escola.db") #caso não exista o banco de dados, será criado um novo para criar a conexão
```

```python
cur = con.cursor() #cria um cursor que permite percorrer os registros do banco
```

```python
sql_create = "create table cursos (id integer primary key, titulo varchar(100), categoria varchar(140))"
#atribui ao objeto "sql_create" o comando SQL para criar a tabela
```

```python
cur.execute(sql_create) #executa o comando gravado no objeto sql_create para criar a tabela no banco de dados
```

```python
sql_insert = "insert into cursos values (?, ?, ?)" #cria uma sentença para isnerção de valores na tabela
```

```python
#Dados a serem inseridos
recset = [(0, "Física Básica", "Física"), 
          (1, "Cálculo", "Matemática"),
          (2, "Introdução ao Python", "Computação")]
```

```python
#Gravação dos dados
for rec in recset:
  cur.execute(sql_insert, rec)
```

```python
#Grava a transação
con.commit()
```

```python
sql_select = "select * from cursos" #cria uma sentença de consulta dos dados
```

```python
cur.execute(sql_select) #executa a query criada em sql_select
dados = cur.fetchall() #salva os registros recuperados no objeto "dados"
```

```python
#Mostrando registros recuperados
for linha in dados:
  print("Curso Id: %d, Título: %s, Categoria: %s \n" % linha)
```

```python
recset = [(3, "Mecânica Quântica", "Física"),(4, "BigData", "Computação")]

for rec in recset:
  cur.execute(sql_insert, rec)

con.commit()
```

```python
cur.execute(sql_select)

recset = cur.fetchall()

for rec in recset:
  print("Curso Id: %d, Título: %s, Categoria: %s \n" % rec)
```

```python
con.close() #fecha a conexão com o banco de dados
```
